Percent of people who don't have a vehicle





```python
rqdata=gpd.read_file('data/w4.geojson')
```

```python
rqdata.info()
```

```python
#columns to keep
columnskeep0 = ['County',
               'B08201_calc_pctNoVehE',
                'geometry'
                 ]
# add it to a new dataframe
rqheatwave= rqdata[columnskeep0]
rqheatwave
```

```python
#B08201_calc_pctNoVehE
#This name is too long to remember, so I want to change the name of it
# "Percent of people don't have a vehicle"  我需要使用双引号, 因为don't 里面有单引号，python无法识别我这个了,需要添加""让他知道里面是String

rqheatwave.columns = ['County',
                      "Percent of people do not have a vehicle",
                      'geometry'
]
rqheatwave
```

```python
# or I can use rename to change to column name
#rqheatwave = rqheatwave.rename(columns={"Percent of people don't have a vehicle": 'pct'})
#rqheatwave
```

```python
rqheatwave.plot(figsize=(12,10),
                 column='Percent of people do not have a vehicle',
                 legend=True, 
                 scheme='NaturalBreaks')
```
